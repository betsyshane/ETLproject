```python
# URL of page to be scraped
url = 'https://screenrant.com/most-anticipated-movies-2017/'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object
soup = BeautifulSoup(response.text, 'html')
```

```python
#scrape anticipated list for movie titles
movie_titles=[]
results= soup.find_all('div', class_= 'w-content')

for result in results:
    titles= result.find_all('h2')
for title in titles:
    name= title.text
    movie_titles.append(name)
movie_titles
```

```python
columns= {"Title": movie_titles}
```

```python
movie_df= pd.DataFrame(columns)
movie_df
```

```python
movie_df["rank"], movie_df["title"] = movie_df["Title"].str.split(' ', 1).str
movie_df_final= movie_df[["rank", "title"]]
movie_df_final
```
