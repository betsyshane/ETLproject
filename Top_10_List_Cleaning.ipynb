```python
# importing the list of 2017 movies
csv_file = "Resources/IMDB Links.csv"
ttmovie_list_df= pd.read_csv(csv_file)
ttmovie_list_df.head()
```

```python
#turning the movies into a list
movielist = list(ttmovie_list_df["MovieName"])
```

```python
# new dataframe for clean data
topten_appearances = {
    "Movie": movielist,
    "Ten": tens_count,
    "Nine": nines_count,
    "Eight": eights_count,
    "Seven": sevens_count,
    "Six": sixs_count,
    "Five": fives_count,
    "Four": fours_count,
    "Three": threes_count,
    "Two": twos_count,
    "One_": ones_count
}

toptencounts_df = pd.DataFrame(topten_appearances)
toptencounts_df.head()
```

```python
# creating a rank score where a number one ranking garners 10 points, two is 9 and so on...then adding together to determine the critical "best" of the year

toptencounts_df["Rank Score"] = (toptencounts_df["Ten"]*1) + (toptencounts_df["Nine"]*2) + \
    (toptencounts_df["Eight"]*3) + (toptencounts_df["Seven"]*4) + \
    (toptencounts_df["Six"]*5) + (toptencounts_df["Five"]*6) + \
    (toptencounts_df["Four"]*7) + (toptencounts_df["Three"]*8) + \
    (toptencounts_df["Two"]*9) + (toptencounts_df["One_"]*10) 

toptencounts_df.head()
```
